```python
# Necessary imports
import numpy as np
from ipywidgets import interact
import matplotlib.pyplot as plt
from bokeh.plotting import figure, show, output_notebook
from bokeh.layouts import gridplot
```

```python
%matplotlib inline
np.random.seed(8)
X = np.random.randn(300,1)
y = 2*(X**3) + 10 + 4.6*np.random.randn(300,1)
plt.plot(X,y,'b.')
plt.title("Random data")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
def peso(x0,X,tau): 
    w= np.exp(np.sum((X - x0) ** 2,axis=1) / (-2 * (tau **2) ))
    return w
```

```python
print(X.shape)
l=0
w=peso(l,np.sort(X,axis=1),0.08)
plt.plot(w)
w=peso(l,X,0.08)
```

```python
theta = np.array([[-10],[-10]])
y=y.reshape((len(y), 1))
X_b = np.c_[np.ones((len(X),1)),X]
X_bs=X_b*1
nc=X_b.shape[1]
for it in range(1,nc):
    m=np.mean(X_b[:,it])
    std=np.std(X_b[:,it])
    X_bs[:,it]=(X_b[:,it]-m)/std

n_iter = 500    
j_cost= np.zeros([n_iter])
ep=1e-4
lr =1e-4
# xw=np.zeros((len(X_bs),1))
for i in range(1):
    x0=X_bs[0,1].reshape((1,))
    xt=X_bs[:,1].reshape(len(X_bs[:,1]),1)
    xw=peso(x0,xt,0.08)
    pred=xw* np.dot(X_bs,theta)
    h=pred - y
    grad= X_bs.T.dot(h)
    theta = theta -lr*grad
    j_cost[i]=np.sum(np.square(h))

fig = plt.figure()
plt.plot(j_cost)
plt.show()
```
