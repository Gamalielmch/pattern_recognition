```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use(['ggplot'])
```

```python
X = 2 * np.random.rand(100,1)
y = 4 +3 * X+np.random.randn(100,1)
```

```python
plt.plot(X,y,'b.')
plt.xlabel("$x$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
_ =plt.axis([0,2,0,15])
```

Minimos cuadrados

$\Large{y=mx+b}$

$\Large{m= \frac{N \sum_{i=1}^{N} {x_iy_i}- \sum_{i=1}^{N}{y_i} \sum_{i=1}^{N}{x_i}}{N \sum_{i=1}^{N}{x_i^2} - (\sum_{i=1}^{N}{x_i})^2}}$

$\Large{b=   \frac{\sum_{i=1}^{N}{y_i} \sum_{i=1}^{N}{x_i^2}- \sum_{i=1}^{N} {x_iy_i}\sum_{i=1}^{N}{x_i}} {N \sum_{i=1}^{N}{x_i^2}-(\sum_{i=1}^{N}{x_i})^2  }}$


```python
N=len(X)
sumxy=np.dot(np.transpose(X),y)
sumy=np.sum(y)
sumx=np.sum(X)
sumx2=sumx**2
sumxx=np.sum(X**2)
m= (N*sumxy - sumy*sumx)/(N*sumxx- sumx2)
b=(sumy*sumxx-sumxy*sumx)/(N*sumxx- sumx2)

b=np.asarray(b)
m=np.asarray(m)
print(m,b)
```

```python
X_new=[0,2]
y_predict=m[0]*X_new+b[0]
print(X_new,y_predict)
```

```python
plt.plot(X_new,y_predict,'r-')
plt.plot(X,y,'b.')
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([0,2,0,15])
```
